```python
# 1. Start a loop that keeps the whole program running
while True:
    print("\n--- Multiplication Table Generator ---")
    
    # 2. Ask the user for a number
    # We use int() because the input comes in as text, and we need a number to do math
    num = int(input("Enter any number to see its times table: "))
    
    print(f"\nTimes Table for {num}:")
    
    # 3. Use a for loop to count from 1 to 12
    # range(1, 13) means "start at 1 and stop BEFORE 13"
    for i in range(1, 13):
        result = num * i
        print(f"{num} x {i} = {result}")
    
    # 4. Ask if they want to go again
    again = input("\nDo you want to try another number? (yes/no): ").strip().lower()
    
    # 5. Check the answer
    if again != 'yes':
        print("Goodbye! Happy coding.")
        break # This exits the while loop
```
