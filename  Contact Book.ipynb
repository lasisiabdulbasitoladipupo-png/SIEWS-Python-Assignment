```python
# --- STEP 1: INITIALIZE THE CONTACT BOOK ---
# No imports needed! Python knows how to handle dictionaries naturally.
contact_book = {}

print("--- Welcome to your Python Contact Book ---")

# --- STEP 2: START THE PROGRAM LOOP ---
while True:
    print("\nMain Menu:")
    print("1. Add a Contact")
    print("2. Search for a Contact")
    print("3. Quit")
    
    # Get the user's choice
    choice = input("Choose an option (1, 2, or 3): ")

    if choice == '1':
        # --- OPTION 1: ADD ---
        name = input("Enter the person's name: ").strip()
        phone = input(f"Enter the phone number for {name}: ").strip()
        contact_book[name] = phone
        print(f"Success! {name} has been added to your book.")

    elif choice == '2':
        # --- OPTION 2: SEARCH ---
        search_name = input("Who are you looking for? ").strip()
        if search_name in contact_book:
            print(f"Found! {search_name}'s number is: {contact_book[search_name]}")
        else:
            print(f"Sorry, '{search_name}' is not in your contacts.")

    elif choice == '3':
        # --- OPTION 3: QUIT ---
        print("Closing the program... Goodbye!")
        break  # This stops the loop immediately

    else:
        # Handle cases where the user types something else (like '4' or 'hello')
        print("Invalid choice. Please enter 1, 2, or 3.")
```
