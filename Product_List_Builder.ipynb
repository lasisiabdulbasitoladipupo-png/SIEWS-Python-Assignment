```python
# 1. Start with an empty list to store the products
products = []

print("--- Welcome to your Product List Builder ---")
print("(Type 'done' when you are finished adding items)")

# 2. Start a loop that keeps asking for items
while True:
    # Get user input
    item = input("Enter a product: ").strip()
    
    # 3. Check if the user wants to stop
    if item.lower() == 'done':
        break  # Exit the loop
    
    # 4. If they didn't type 'done', add the item to the list
    products.append(item)
    print(f"Added: {item}")

# 5. At the end, print the full list
print("\n--- Your Final Product List ---")
if len(products) > 0:
    for product in products:
        print(f"- {product}")
else:
    print("The list is empty!")
```
