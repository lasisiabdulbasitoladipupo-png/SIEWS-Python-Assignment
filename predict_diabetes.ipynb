```python
# ── Helper function with input validation ──
def get_input(prompt, example):
    while True:
        try:
            return float(input(f'{prompt} [e.g. {example}]: '))
        except ValueError:
            print('  Please enter a number.')
```

```python
# ── Collect patient data ──
pregnancies = get_input('Pregnancies',             '2')
glucose     = get_input('Glucose (mg/dL)',          '120')
bp          = get_input('Blood Pressure (mm Hg)',   '70')
skin        = get_input('Skin Thickness (mm)',       '20')
insulin     = get_input('Insulin (mu U/ml)',         '80')
bmi         = get_input('BMI (kg/m²)',               '25.0')
dpf         = get_input('Diabetes Pedigree Func',    '0.5')
age         = get_input('Age (years)',               '35')
```

```python
import numpy as np
import pandas as pd
import joblib

# ── Load saved scaler and model ──
model  = joblib.load('diabetes_model.pkl')
scaler = joblib.load('diabetes_scaler.pkl')

# ── Prepare input as DataFrame (matches scaler's expected format) ──
data = pd.DataFrame([[pregnancies, glucose, bp, skin, insulin, bmi, dpf, age]],
                    columns=['Pregnancies','Glucose','BloodPressure','SkinThickness',
                             'Insulin','BMI','DiabetesPedigreeFunction','Age'])

# ── Scale and predict ──
scaled = scaler.transform(data)
pred   = model.predict(scaled)[0]
proba  = model.predict_proba(scaled)[0]
```

```python
# ── Display Result ──
print('\n' + '='*50)
if pred == 1:
    print(f'  RESULT     : ⚠️  DIABETIC (Positive)')
    print(f'  CONFIDENCE : {proba[1]*100:.1f}% likely diabetic')
    print('\n  RECOMMENDATION: Please consult a doctor')
    print('  immediately for proper diagnosis.')
else:
    print(f'  RESULT     : ✅  NOT DIABETIC (Negative)')
    print(f'  CONFIDENCE : {proba[0]*100:.1f}% likely healthy')
    print('\n  RECOMMENDATION: Maintain healthy lifestyle.')
    print('  Regular checkups recommended.')

print('='*50)
print(f'\nProbabilities:')
print(f'  Not Diabetic : {proba[0]*100:.1f}%')
print(f'     Diabetic  : {proba[1]*100:.1f}%')
print('\nNOTE: ML prediction only — not medical advice.')
print('Always consult a qualified healthcare professional.\n')
```
