```python
# write a program to create a contact book
books = []
num = int(input('How many books do you want to add?'))
for n in range(num):
    print(f'Enter book {n+1}')
    bookName = input('Enter book name')
    phoneNumber = input('Enter phone number: ')
    book = {
        'name': bookName,
        'phone': phoneNumber
    }
    books.append(book)
print('Allbooks are added')
```
